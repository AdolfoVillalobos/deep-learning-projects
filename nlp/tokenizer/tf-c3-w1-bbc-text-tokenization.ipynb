```python
def read_csv(path):
    sentences = []
    labels = []
    with open(path, "r") as csvfile:
        for i, row in enumerate(csvfile.readlines()):
            if i>0:
                line = row.split(",")
                category = line[0]
                corpus = line[1]
                for word in stopwords:
                    token = " "+word+" "
                    corpus = corpus.replace(token, " ")
                    corpus = corpus.replace("  "," ")
                labels.append(category)
                sentences.append(corpus)
    return sentences, labels   
```

```python
sentences, labels = read_csv(BASE_DIR)
```

## Tokenize Data

```python
def tokens(corpus, oov_token="<OOV>"):
    
    # Fit the Tokenizer
    tokenizer = Tokenizer(oov_token=oov_token)
    tokenizer.fit_on_texts(corpus)
    word_index = tokenizer.word_index
    
    # Transform
    sequences = tokenizer.texts_to_sequences(corpus)
    
    #Padding
    padded = pad_sequences(sequences, padding="post")
    
    print(f"Vocabulary Length: {len(word_index)}")
    print(f"Shape of Padded Sentences: {padded.shape}")
    
    return padded, word_index, tokenizer
    
    
    
    
```

```python
# Fit the Tokenizer

sequences, word_index, sentence_tokenizer = tokens(sentences)
```

```python
labels_seq, labels_index, labels_tokenizer = tokens(labels, oov_token=None)
```
